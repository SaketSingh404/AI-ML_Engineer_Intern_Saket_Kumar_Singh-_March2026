## Loading  PDF 

```python
loader = PyMuPDFLoader('../Documents/BTech_CreditSystem_NEP-ocr.pdf')
```

```python
docs = loader.load()
```

## Splitting

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)
```

```python
chunks = []

for i in range(0,len(docs)):
    text = docs[i].page_content
    chunk = splitter.split_text(text)
    # print(text)
    for j in chunk:
        chunks.append(j)
```

```python
chunks_documents = splitter.create_documents(chunks)
```

## Making Embeddings

```python
embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en"
)
```

```python
vector_store = FAISS.from_documents(chunks_documents, embeddings)
```

```python
vector_store.index_to_docstore_id
```

## Retriever

```python
retriever = vector_store.as_retriever(search_kwargs={'k':4})
```

```python
query = "What courses I can take in applied physics department"
result = retriever.invoke(query)
for i, doc in enumerate(result):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```
