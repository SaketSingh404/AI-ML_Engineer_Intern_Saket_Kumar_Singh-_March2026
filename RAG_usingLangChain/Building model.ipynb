## Loading  PDF 

```python
loader = PyMuPDFLoader('../Documents/BTech_CreditSystem_NEP-ocr.pdf')
```

```python
docs = loader.load()
```

## Splitting

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)
```

```python
# chunks = []

# for i in range(0,len(docs)):
#     text = docs[i].page_content
#     chunk = splitter.split_text(text)
#     # print(text)
#     for j in chunk:
#         chunks.append(j)
```

```python
chunks_documents = splitter.split_documents(docs)
```

## Making Embeddings

```python
embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en"
)
```

```python
vector_store = FAISS.from_documents(chunks_documents, embeddings)
```

```python
# vector_store.index_to_docstore_id
```

## Retriever

```python
retriever = vector_store.as_retriever(search_kwargs={'k':4})
```

```python
query = "What courses I can take in applied physics"
result = retriever.invoke(query)
for i, doc in enumerate(result):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```

```python
print(f"Source = {result[2].metadata['source']}\nPage_number={str(result[2].metadata['page'])}")
```

```python
vector_store.save_local('credit_system')
```

```python
all_docs = []

folder_path = "../Documents/"
```

```python
os.listdir(folder_path)
```

```python
for file in os.listdir(folder_path):
    if file.endswith(".pdf"):
        loader = PyMuPDFLoader(os.path.join(folder_path, file))
        docs = loader.load()

        for d in docs:
            d.metadata["source"] = file
            d.metadata["department"] = file.split("_")[0]

        all_docs.extend(docs)
```

```python
db = FAISS.from_documents(all_docs, embeddings)
```

```python
db.save_local("../vector_database")
```
